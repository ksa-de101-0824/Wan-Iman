```python
import requests
import json

def get_salary_data():
    url = 'https://raw.githubusercontent.com/nextynewb/datasets/main/salaries.json'
    response = requests.get(url)
    response.raise_for_status()
    salaries_data = response.json()
    return salaries_data['data']


salaries = get_salary_data()
print(salaries)

usd_rate = 4.2

for salary in salaries:
    new_salary = round(float(salary["pay"].replace('$','').replace('K','')) * 1000, 2)
    salary['pay_usd'] = new_salary 
    salary['pay_myr'] = round(salary['pay_usd'] * usd_rate, 2 )
    pay_myr = salary['pay_myr']

    if pay_myr < 100000:
        income_class = 'low'
    elif pay_myr <= 200000:
        income_class = 'medium'
    else:
        income_class = "high"

    salary['income class'] = income_class

    print(salary)
```
