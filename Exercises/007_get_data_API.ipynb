```python
import requests

def get_news_data(offset):
    offset = offset * 10
    url = f"https://theedgemalaysia.com/api/loadMoreCategories"
    parameters = {
        'offset': offset, # This is the starting index for fetching the news data
        'categories': 'malaysia'
    }
    response = requests.get(url, params=parameters)
    data = response.json()
    results = data['results']
    return results
```

```python
articles = []

def data():
    num_api_calls = 10

    for i in range(num_api_calls):
        offset = i
        news_data = get_news_data(offset) 
        print(news_data) # Print each data 
        articles.extend(news_data)

    return articles

articles_data = data()
print("++++++++++++++++++++++++")
print(f"Total of data: {len(articles_data)}")
print("++++++++++++++++++++++++")
print(articles) #Print all data in one list 

```
